# Extended Universe Data Pull

Download ETFs, precious metals, and individual stock data for strategy research.

## Code Steps

1. Define an extended universe: market ETFs, gold/silver ETFs, and large individual stocks.
2. Download daily adjusted price data from yfinance.
3. Save the raw yfinance CSV in `data/raw/`.
4. Extract close prices and save a cleaner `extended_close_prices.csv` file in `data/processed/`.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.research.data import EXTENDED_UNIVERSE, download_yfinance_prices, load_yfinance_close

RAW_DIR = PROJECT_ROOT / "data" / "raw"
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
```

```python
tickers = EXTENDED_UNIVERSE
start = "2015-01-01"
interval = "1d"

tickers
```

```python
raw_path = download_yfinance_prices(
    tickers,
    start=start,
    interval=interval,
    output_dir=RAW_DIR,
    threads=False,
)

raw_path
```

```python
close = load_yfinance_close(raw_path, tickers=tickers)
close_path = PROCESSED_DIR / "extended_close_prices.csv"
close.to_csv(close_path)

close.shape, close.index.min(), close.index.max(), close_path
```
